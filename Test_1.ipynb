```python
# 5.	Определение «плохой»/ «хороший» актив (1/0)

for i in input.index:
    if input['Наименование актива'][i] in ['Займ страхователю по договору страхования жизни']:
        input['Маркер (1)/(0)'][i] = 1
    
    elif input['Наименование актива'][i] in ['Вексели','Вклад в капитал','Нематериальные активы','Основные средства (кроме недвижимого имущества)','Отложенные аквизиционные расходы','Отложенные налоговые активы','Прочие активы (не торгуемые товары)','Слитки  золота,  серебра, платины и палладия, а также памятные монеты Российской Федерации из драгоценных металлов']:
        input['Маркер (1)/(0)'][i] = 0
    
    elif input['Наименование актива'][i] == 'Денежные средства':
        if input['Комментарий по активу'][i] != 'денежные средства в пути' and pd.isnull(input['Кредитный рейтинг  контрагента'][i]) == False:
            input['Маркер (1)/(0)'][i] = 1
        else:
            input['Маркер (1)/(0)'][i] = 0
    
    elif input['Наименование актива'][i] == 'Депозиты':
        if input['Комментарий по активу'][i] != 'субординированный депозит':
            input['Маркер (1)/(0)'][i] = 1
        else:
            input['Маркер (1)/(0)'][i] = 0
            
    elif input['Наименование актива'][i] == 'Недвижимость':
        if input['Комментарий по активу'][i] != 'земля, не образует единый компелекс со зданием':
            input['Маркер (1)/(0)'][i] = 1
        else:
            input['Маркер (1)/(0)'][i] = 0
            
    elif input['Наименование актива'][i] == 'Производный финансовый инструмент':
        if input['Комментарий по активу'][i] == 'форвард на валюту':
            input['Маркер (1)/(0)'][i] = 1
        else:
            input['Маркер (1)/(0)'][i] = 0
            
    elif input['Наименование актива'][i] == 'Требования по текущему налогу на прибыль':
        if input['Комментарий по активу'][i] == 'Можно востребовать деньгами':
            input['Маркер (1)/(0)'][i] = 1
        else:
            input['Маркер (1)/(0)'][i] = 0
            
    elif input['Наименование актива'][i] == 'Ценные бумаги':
        if input['Комментарий по активу'][i] not in ['субординированная облигация, не конвертируемая в акции', 'акция, доля владения >10%', 'ипотечный сертификат участия']:
            input['Маркер (1)/(0)'][i] = 1
        else:
            input['Маркер (1)/(0)'][i] = 0

    elif input['Наименование актива'][i] in ['Займ','Доля перестраховщика в страховых резервах (регуляторная)']:
        if pd.isnull(input['Кредитный рейтинг  контрагента'][i]) == False:
            input['Маркер (1)/(0)'][i] = 1
        else:
            input['Маркер (1)/(0)'][i] = 0
            
    elif input['Наименование актива'][i] == 'Дебиторская задолженность':
        if pd.isnull(input['Кредитный рейтинг  контрагента'][i]) == False or pd.isnull(input['Период погашения дебиторской задолженности не превышает 10 рабочих дней'][i]) == False:
            input['Маркер (1)/(0)'][i] = 1
        else:
            input['Маркер (1)/(0)'][i] = 0
            
    else:
        input['Маркер (1)/(0)'][i] = 2
           
print(input['Маркер (1)/(0)'].sample(10))
```
